```python
#fib series on tree recursion 0 1 1 2 3 5 8
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
terms=int(input("enter number of terms to print:"))
print("fibonocci series...")
for i in range(terms):
    print(fib(i), end= ' ')
```

```python
'''permutations of char in tree recursion'''
def permute(s, bucket=' '):
    if not s:
        print(bucket)
        return
    for i in range(len(s)):
        ns=s[:i]+s[i+1:]
        permute(ns, bucket+s[i])
text=input("enter a name/word: ")
print("possibilities of combination......")
permute(text)
```

```python
'''permutations of binary in tree recursion'''
def binary(n, b=' '):
    if n==0:
        print(b)
        return
    binary(n-1, b+'0')
    binary(n-1, b+'1')
length=int(input("enter a name/word: "))
print("binary combination......")
binary(length)
```

```python
#head recursion
def head(n):
    if n==0:
        return
    head(n-1)
    print(n)
num=int(input("enter a number:"))
head(num)        
```

```python
#tail recorion
def tail(n):
    if n==0:
        return
    print(n)
    tail(n-1)
num=int(input("enter a number:"))
tail(num)        
```

```python
def head(str):
    if str==0:
        return
    print(str)
    head(str)
string=input("enter a name:")
head(string) 
```

```python
#head recusion with sum
def sumhead(n):
    if n==0:
        return 0
    return n+sumhead(n-1)
num=int(input("enter a number:"))
print("sum:",sumhead(num))
```

```python
#tail recusion with sum
def sumtail(n, temp=0):
    if n==0:
        return temp
    return n+sumtail(n-1 , temp+n)
num=int(input("enter a number:"))
print("sum:",sumtail(num))
```
